# Angular Template Tag Debugger
This notebook analyzes `page3.component.html` for mismatched `<div>` and `<ng-container>` tags.

```python
from pathlib import Path
import re

path = Path(r'c:\Users\fo.technicalsupport\Desktop\Projects\angular\daily-requisition\src\app\dashboard\pages\page3\page3.component.html')
text = path.read_text(encoding='utf-8')

# Find tag occurrences
regex = re.compile(r'<(/?)(ng-container|div)([^>]*)>', re.IGNORECASE)
stack = []
errors = []
for m in regex.finditer(text):
    tag = m.group(2).lower()
    closing = bool(m.group(1))
    line = text.count('\n', 0, m.start()) + 1
    if not closing:
        stack.append((tag, line, m.group(0)))
    else:
        if stack and stack[-1][0] == tag:
            stack.pop()
        else:
            errors.append((tag, line, m.group(0), stack[-10:]))
            # attempt to recover
            for idx in range(len(stack)-1, -1, -1):
                if stack[idx][0] == tag:
                    stack.pop(idx)
                    break

print('Open div count:', len(re.findall(r'<div[^>]*>', text)))
print('Close div count:', len(re.findall(r'</div>', text)))
print('Open ng-container count:', len(re.findall(r'<ng-container[^>]*>', text)))
print('Close ng-container count:', len(re.findall(r'</ng-container>', text)))
print('Errors:', len(errors))
for err in errors[:10]:
    print(err)
print('Remaining stack length:', len(stack))
for item in stack[-20:]:
    print('Remaining open', item)
```
